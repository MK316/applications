* Description: This application generates an audio file of words you select from the list.

#  [1] Install and import required libraries

```python
#@markdown Install and import required libraries
%%capture
!pip install gTTS
from gtts import gTTS
import ipywidgets as widgets
import pandas as pd
from IPython.display import display, Audio
from ipywidgets import Button, HBox, Output
from gtts import gTTS
import requests
import io
```

#  [2] How to use: step by step

+ Step 1: Run the code below.()
+ Step 2: You'll see two box prompts. Type starting (e.g., 1) and ending index (e.g., 100) referring to the Oxoford 3K wordlist.
+ Step 3: Wait until the audio is generated and displayed on the screen.

To download the audio file: 
+ click the icon on the left panel (a folder shape icon), 
+ find the audio file: the file name is in the format start_number_to_end_number.mp3, based on the numbers you've entered., 
+ click on the three dots at the end of the file name to access the download menu.

```python
#@markdown Creates an audio file with the selected range of words from the 3K wordlist.
# 1. Read a dataframe from a specific URL
url = "https://raw.githubusercontent.com/MK316/workshop22/main/data/oxford3k.csv"
df = pd.read_csv(url)

def on_button_click(button):
    x = start_num_widget.value
    y = end_num_widget.value

    words = df["WORD"][x - 1:y].tolist()
    wordlist = ", ".join(words)
    text_to_say = f"Let me say words from {x} to {y}: " + wordlist + "!" + "This is the end."
    audio_filename = f"{x}_to_{y}.mp3"
    generate_audio(text_to_say, audio_filename)

    with audio_output:
        audio_output.clear_output()
        display(Audio(audio_filename))

def generate_audio(text, filename):
    tts = gTTS(text)
    tts.save(filename)

start_num_widget = widgets.IntText(description='Starting Index:', value=1)
end_num_widget = widgets.IntText(description='Ending Index:', value=100)
button_widget = Button(description='Submit')
audio_output = Output()

button_widget.on_click(on_button_click)

display(HBox([start_num_widget, end_num_widget, button_widget]))
display(audio_output)
```
