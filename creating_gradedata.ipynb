# Popular names and creating sample data

* Data: Popular names in 2021, top 1,000
[source](https://www.ssa.gov/oact/babynames/)

```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/MK316/data/main/names1000.csv"
f = pd.read_csv(url)
print(f.head())
f.tail()
```

```python
# Add [gender] column

import numpy as np

x = np.array(["Male", "Female"])
y = np.repeat(x, [500, 500], axis=0)

f['Gender'] = y
f.head()
```

```python
data = f.sample(120)
len(data)
```

```python
# 3 classes, ID = 1~40
x = ["Class_A","Class_B","Class_C"]
z = np.repeat(x, [40, 40, 40], axis=0)
id = list(range(1,41))
id = id*3

data['ID'] = id
data['Class'] = z
len(data); data.head()
```

## Creating random data with {Numpy} package

e.g., Gradebook: 3 classes with 40 students per class, 3 subjects

*Goal: create random grades of 3 subjects (English, Korean, Math)

```python
import numpy as np
import random
```

```python
#@markdown Create dataframe with random grades for a subject: 

#@markdown Todo: Write the name of the subject in a popup window. (e.g., Korean, English, Math)


eg = np.random.normal(58, 20, 10000)
kg = np.random.normal(62, 10, 10000)
mg = np.random.normal(52, 30, 10000)

eg1 = [s for s in eg if 20 < s < 100.1]
kg1 = [s for s in kg if 40 < s < 100.1]
mg1 = [s for s in mg if 30 < s < 100.1]

eg2 = []
for i in range(0,len(eg1)):
  m = round(eg1[i])
  eg2.append(m)

kg2 = []
for i in range(0,len(kg1)):
  m = round(kg1[i])
  kg2.append(m)

mg2 = []
for i in range(0,len(mg1)):
  m = round(mg1[i])
  mg2.append(m)

# Random sampling of 120 for each subject

kgrade = random.sample(kg2,120)
egrade = random.sample(eg2,120)
mgrade = random.sample(mg2,120)

# subj = input()
# subject = str(subj)

data['Korean'] = kgrade
data['English'] = egrade
data['Math'] = mgrade
```
