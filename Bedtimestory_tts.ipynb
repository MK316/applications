# 🌀**Step 2: Set up TTS**

```python
#@markdown 🚩 {gTTS} package installation and import
%%capture
!pip install gTTS
from gtts import gTTS
from IPython.display import Audio
```

```python
#@markdown 🚩 Getting ready to use { tts ( _text_to_say_) }:
def tts(mytext):

#Install gTTS
# !pip install gTTS
# from gtts import gTTS
# from IPython.display import Audio

  text_to_say = mytext

# Step ⓵ Language to choose:
  language_to_choose = "en" #@param ["en", "fr","ko",'es']
  Slow_speed = "False" #@param ["True","False"]
  # lang = language_to_choose

  print("Play language accent: %s"%language_to_choose)
  language = language_to_choose

# gTTS
  gtts_object = gTTS(text = text_to_say,
                     lang = language,
                    slow = Slow_speed)
  
#@markdown Filename to save:
  print("Write audio file name to save: type e.g) 'myaudio' => The audio file will be saved as 'myaudio.wav'")
  audiofilename = input()
  aname = audiofilename + '.wav'

# #@markdown Step ③: Create the audio file (.wav) to play:

  gtts_object.save(aname)

# # Output
  return Audio(aname)
```

# 🌀**Step 3: Creating audio files** 

After running the code, the audio file will be saved on the left panel. (Alternativel, you can download the audio from the audio box, click the vertical three dots in the audio box)

```python
#@markdown 📌 The audio file is saved as **"myaudio.wav"**
print(text)
tts(text)
```
