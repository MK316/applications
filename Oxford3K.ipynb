```python
#@markdown ðŸ”³ Data: Oxford learner's vocabulary 3K: df, show first 5, last 5 words
import pandas as pd

url = "https://raw.githubusercontent.com/MK316/workshop22/main/data/oxford3k.csv"

df = pd.read_csv(url)
print(df.head())
print(df.tail())
```

# ðŸŒ€**Step 2: Set up TTS**

```python
#@markdown ðŸš© {gTTS} package installation and import
%%capture
!pip install gTTS
from gtts import gTTS
from IPython.display import Audio
```

```python
#@markdown ðŸš© Getting ready to use { tts ( _text_to_say_) }:
def tts(mytext):

#Install gTTS
# !pip install gTTS
# from gtts import gTTS
# from IPython.display import Audio

  text_to_say = mytext

# Step â“µ Language to choose:
  language_to_choose = "en" #@param ["en", "fr","ko",'es']
#  Slow_speed = "True" #@param ["True","False"]
  # lang = language_to_choose

  print("Play language accent: %s"%language_to_choose)
  language = language_to_choose

# gTTS
  gtts_object = gTTS(text = text_to_say,
                     lang = language,
                    slow = False)
  
# #@markdown Step â‘¢: Create the audio file (.wav) to play:
  gtts_object.save("wordlist.mp3")

# # Output
  return Audio("wordlist.mp3")
```
