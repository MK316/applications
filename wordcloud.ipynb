# Wordcloud plot 1: with dark background


```python
#@markdown ◘ Create a wordcloud plot with the given text:
wc = WordCloud().generate(text) 
plt.imshow(wc)
# plt.axis("off")
```

```python
#@markdown ◘ Run this if you want to remove the axis marks in the plot:
wc = WordCloud().generate(text) 
plt.imshow(wc)
plt.axis("off")
```

```python
#@markdown ◘ Save the wordcloud plot on colab (left panel): 'wordcloud1.png'

wc.to_file('wordcloud1.png')
imgfile = 'wordcloud1.png'
print('Image file is saved as %s. Check Files on the left.'%imgfile)
```

# Wordcloud 2: with white background

```python
#@markdown ◘ Remove stop words
stopwords = set(STOPWORDS)
#stopwords.add('us')
len(stopwords)
spltxt = text.split()
```

```python
#@markdown ◘ Wordcloud plot:
wordcloud = WordCloud(stopwords=stopwords,relative_scaling = 0.2, random_state=3,
                    # max_words=2000, background_color='white').generate(' '.join(spltxt))
                      max_words=2000, background_color='white').generate(' '.join(spltxt))
plt.figure(figsize=(12,12))
plt.imshow(wordcloud, interpolation="bilinear")
wordcloud.to_file('wordcloud_2.png')

print('Wordcloud file is saved in the left panel: wordcloud2.png')
```

```python
#@markdown ◘ Wordcloud plot (without axis marks):
wordcloud = WordCloud(stopwords=stopwords,relative_scaling = 0.2, random_state=3,
                    # max_words=2000, background_color='white').generate(' '.join(spltxt))
                      max_words=2000, background_color='white').generate(' '.join(spltxt))
plt.figure(figsize=(12,12))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
wordcloud.to_file('wordcloud_2.png')

print('Wordcloud file is saved in the left panel: wordcloud2.png')

```
