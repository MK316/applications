```python
#@markdown 2) Removing time stamps, e.g., "01:32"
# Remove Time stamps (01:09) > The result is assigned back to the variable named "text".
import re

t1 = text.split(" ")
numbers = re.findall('[0-9]+:[0-9]+',text)
t2 = [w for w in t1 if w not in numbers]
text1 = " ".join(t2)
text = text1
text

print("ðŸ‘Œ Okay, your text is ready for analysis.")
```
